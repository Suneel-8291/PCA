```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
```

```python
data = load_breast_cancer()
```

```python
df = pd.DataFrame(data.data,columns = data.feature_names)
```

```python
df
```

```python
std_scale = StandardScaler()
```

```python
std_scale.fit(df)
```

```python
scaled_data = std_scale.transform(df)
```

```python
scaled_data
```

```python
principle_component_analysis = PCA(n_components = 2)
```

```python
principle_component_analysis.fit(scaled_data)
```

```python
x_pca = principle_component_analysis.transform(scaled_data)
```

```python
x_pca
```
